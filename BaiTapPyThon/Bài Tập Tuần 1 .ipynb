# Thử Tô Màu

* Bạn Bình mua được một tờ giấy kẻ ô được tô màu trắng đen lẫn lộn. Nhìn tờ giấy tô màu không có quy tắc gì cả, Bình bỗng nghĩ ra một câu hỏi thú vị. Giả sử chúng ta chọn một vài hàng và một vài cột nào đó (có thể không chọn cũng được), sau đó tô lại tất cả các ô trong những hàng và cột được chọn thành màu đỏ thì có cách nào để số ô màu đen còn lại sẽ đúng bằng số tuổi của Bình hay không?

- Hãy viết chương trình giúp Bình đếm xem có bao nhiêu cách chọn hàng và cột thỏa mãn yêu cầu.

 **INPUT**
* Dòng đầu tin chứa 03 số nguyên H, W, K, lần lượt là số hàng, số cột và số tuổi của Bình.

* H dòng tiếp theo sẽ có đúng W ký tự viết liền với nhau. Ký tự dấu chấm (".") đại diện cho ô màu trắng, và ký tự dấu hashtag ("#") đại diện cho ô màu đen.

* Cho biết H và W sẽ không vượt quá 6 và Bình thì không quá H*W tuổi

 **OUTPUT**
* Số cách lựa chọn thỏa mãn yêu cầu

 **Ví dụ**

| Input     | Output    | Giải thích                                       |
|-----------|-----------|--------------------------------------------------|
| 2 3 2     |           |                                                  |
| ..#       | 5         |  5 cách lựa chọn tô màu đó là:                  |
| ###       |           | - Hàng 1, cột 1                                 |
|           |           | - Hàng 1, cột 2                                 |
|           |           | - Hàng 1, cột 3                                 |
|           |           | - Cột 1 và cột 2                                |
|           |           | - Chỉ tô cột 2                                   |
|           |           |                                                  |
| 2 3 4     |           |                                                  |
| ..#       | 1         | Chỉ có 1 lựa chọn đúng, là đừng có tô gì cả    |
| ###       |           |                                                  |
| 2 2 3     |           |                                                  |
| ##        | 0         | Tô hay không tô gì thì cũng không đáp ứng yêu   |
| ##        |           | cầu                                              |
|           |           |                                                  |
| 6 6 8     |           |                                                  |
| ..##..    | 208       |                                                  |
| .#..#.    |           |                                                  |
| #....#    |           |                                                  |
| ######    |           |                                                  |
| #....#    |           |                                                  |
| #....#    |           |                                                  |






```python
def count_ways_to_choose_rows_and_columns(H, W, K, grid):
    total_ways = 0

    # Duyệt qua tất cả các tập hợp con của hàng và cột
    for row_mask in range(1 << H):
        for col_mask in range(1 << W):
            black_remaining = 0

            # Đếm số ô đen còn lại sau khi đã tô màu đỏ
            for i in range(H):
                for j in range(W):
                    if (row_mask & (1 << i)) or (col_mask & (1 << j)):
                        continue
                    if grid[i][j] == '#':
                        black_remaining += 1

            # Kiểm tra nếu số ô đen còn lại bằng số tuổi của Bình
            if black_remaining == K:
                total_ways += 1

    return total_ways

# Đọc input
H, W, K = map(int, input().split())
grid = [input().strip() for _ in range(H)]
initial_black_count = sum(row.count('#') for row in grid)
#if initial_black_count == K:
    #print(1)
    #exit()
# Tính và in ra số cách lựa chọn thỏa mãn yêu cầu
print(count_ways_to_choose_rows_and_columns(H, W, K, grid))
```

# Cặp Mã Hóa

* An đang làm việc trong một công ty lớn và có trách nhiệm bảo mật dữ liệu quan trọng của công ty. Để đảm bảo an toàn cho dữ liệu cho công ty, An đã tham gia nhóm phát triển hệ thống mã hóa bất đối xứng với việc tạo cặp khóa mã hóa dựa trên những số “đặc biệt” mà nhóm của An mới tìm ra. Số “đặc biệt" là một số nguyên dương không thể chia hết cho bình phương của bất kỳ số nguyên nào lớn hơn 1. Sự phức tạp của những số “đặc biệt” này rất thích hợp cho việc tạo ra những khóa mã hóa mạnh.

* Bây giờ, nhóm của An muốn biết có bao nhiêu cặp khóa mã hóa mạnh với mỗi khóa nằm trong khoảng số nguyên nhất định từ L đến R. Một cặp khóa mã hóa (x, y) được xem là mạnh nếu cả khóa x, khóa y và giá trị x*y đều là những số “đặc biệt”, thêm vào đó thì x phải luôn nhỏ hơn y. An chịu trách nhiệm chính cho việc viết chương trình để tính số cặp khóa mạnh này nhưng khổ nỗi là cậu ấy đã quên gần hết việc lập trình. Bạn hãy giúp An viết chương trình này nhé!

 là 2 giới hạn của khoảng giá trị của các khóa mã hóa.

**Ouput:**

* Số cặp khóa mã hóa mạnh mà nhóm An muốn biết

**Ví dụ**

| Input | Output |
|-------|--------|
| 3     | 5      |
| 9     |        |

**GIải thích:**
Với L = 3, R = 9: Thì có 5 cặp khóa mã hóa là (3, 5), (3, 7), (5, 6), (5, 7), (6, 7).



```python
def gcd(L,R):
    while R:
        L, R = R, L % R
    return L

def is_special_number(num):
    if int(num**0.5)**2 == num or num % 4 == 0 or num % 9 == 0:
        return 0

    sqrt_num = int(num**0.5) + 1
    for i in range(5, sqrt_num,6):
        if num % i == 0 and (num % i**2 == 0 or int((num / i) ** 0.5) ** 2 == num / i):
          return 0
        elif num % (i + 2) == 0 and ((num % (i + 2)**2 == 0 or int((num / (i + 2)) ** 0.5) ** 2 == num / (i + 2))):
          return 0
    return 1

a, b = map(int, input().split())

b = b + 1
if a == 1:
    numbers = [1]
    for i in range(a + 1, b):
        if is_special_number(i):
            numbers.append(i)
else:
    numbers = list(i for i in range(a, b) if is_special_number(i))

count = 0
length = len(numbers)
for i in range(length):
    for j in range(i+1, length):
        if gcd(numbers[i], numbers[j]) == 1:
            count += 1
print(count)
```

# Đảo Chuỗi

* Bình có một chuỗi a và Bình có thể thực hiện nhiều lần thao tác sau lên chuỗi a: "Chọn một chuỗi con bất kỳ có độ dài lẻ trong a và đảo ngược chuỗi con này"

* Hỏi sau nhiều lần thao tác đó Bình có thể biến chuỗi a thành chuỗi B không?

**INPUT**
* Dòng đầu tiên chứa số t đây là số test case
* 2*t dòng tiếp theo sẽ là các cặp chuỗi a, b
**OUTPUT**

* Ứng vỡi mỗi cặp chuỗi a, b xuất ra trên 1 dòng chữ "YES" hoặc "NO" theo yêu cầu
**Ví dụ**

|       INPUT       | Output |
|:-----------------:|:------:|
|         3         |  YES   |
|       hanoi       |   NO   |
|       ionah       |  YES   |
|         ac        |        |
|         wa        |        |
|        hue        |        |
|        hue        |        |


```python
def same_characters(a, b):
    a , b = list(a), list(b)
    CountA = {}
    CountB = {}
    for i in range(len(a)):
        if a[i] not in b:
            return 0
        CountA[a[i]] = CountA.get(a[i], 0) + 1
        CountB[b[i]] = CountB.get(b[i], 0) + 1
    for char in a:
        if CountA[char] != CountB.get(char, 0):
            return 0
    return 1

def same_characters_at_odd_index(a,b):
    a , b = list(a), list(b)
    CountA = {}
    CountB = {}
    for i in range(0,len(a),2):
        CountA[a[i]] = CountA.get(a[i], 0) + 1
    for i in range(0,len(b),2):
        CountB[b[i]] = CountB.get(b[i], 0) + 1
    for i in range(0,len(a),2):
        if CountA[a[i]] != CountB.get(a[i],0):
            return 0
    return 1
def reversed(a,b):
    if a == b:
        return "YES"
    if len(a) != len(b) or len(a) < 3 or same_characters(a,b) == 0:
        return "NO"
    if same_characters_at_odd_index(a,b) == 1:
        return "YES"
    return "NO"


t = int(input())
lista = []

for _ in range(t):

    a = input().strip()
    b = input().strip()

    lista.append(reversed(a,b))

print(*lista,sep="\n")
```

Các thông báo này nếu có cụm từ "xxx% higher than in-store" trong đó xxx là một số thập phân bất kỳ thì nghĩa là có thể tiết kiệm thêm xxx% giá trị món hàng nếu mua tại cửa hàng. Và nếu có cụm từ "xxx% lower than in-store" thì nghĩa là đã tiết kiệm được xxx% giá trị món hàng nhờ vào việc mua trực tuyến.

* Dòng cuối cùng là số tiền Bình có.

**OUTPUT**
* Nếu Bình mua được hết đống hàng thì xuất ra là true, ngược lại xuất ra false

| Input                                                    | Output |
|----------------------------------------------------------|--------|
| 501700 989900 254900<br>Contratulation that was 2.4% lower than in-store<br>Should reconsider this item is 20.0% higher than in-store<br>Nice, 22.7% lower than in-store<br>3300 | false  |
| 501700 989900 254900<br>Contratulation that was 2.4% lower than in-store<br>Should reconsider this item is 20.0% higher than in-store<br>Nice, 22.7% lower than in-store<br>2000000 | true   |


```python
def extract_percentage(string):
    words = string.split()
    for word in words:
        if "%" in word:
            index = word.find('%')
            substring = word[:index]
            try:
                percentage = float(substring)
                return percentage
            except ValueError:
                pass
    return None

def can_buy_all_items(prices, messages, money):
    for i in range(len(prices)):
        price = prices[i]
        message = messages[i]

        percentage = extract_percentage(message)
        if percentage is not None:
            if "higher than in-store" in message:
                price *= (100 - percentage) / 100
            elif "lower than in-store" in message:
                price *= (100 + percentage) / 100

        if price > money:
            return "false"
        money -= price

    return "true"

if __name__ == "__main__":
    prices_input = input()
    prices = list(map(int, prices_input.split()))

    n = len(prices)

    messages = []
    for i in range(n):
        message = input()
        messages.append(message)

    money = int(input())

    result = can_buy_all_items(prices, messages, money)
    print(result)
```
